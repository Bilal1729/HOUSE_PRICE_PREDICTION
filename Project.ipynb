**4. PREDICTION**

**4.1 LINEAR REGRESSION**

```python
x=df1.drop(['SalePrice'],axis='columns')
y=df1['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
print("Shape of x_train : ",x_train.shape)
print("Shape of x_test : ",x_test.shape)
print("Shape of y_train : ",y_train.shape)
print("Shape of y_test : ",y_test.shape)
```

```python
from sklearn import linear_model
lr1 = linear_model.LinearRegression()
lr1.fit(x_train,y_train)
```

```python
coeffs1 = pd.DataFrame(lr1.coef_,x.columns,columns=['Coefficient'])
coeffs1
```

```python
pred1 = lr1.predict(x_test)
```

```python
scores1 = pd.DataFrame({'Actual':y_test,'Predictions':pred1})
g=sb.pairplot(data=scores1[['Actual','Predictions']],x_vars=['Actual'],y_vars='Predictions')
g.fig.set_figheight(6)
g.fig.set_figwidth(10)
plt.title("Actual VS Prediction (Linear Regression)")
plt.savefig("fig4")
plt.show()
```

```python
from sklearn import metrics
print(y_train.mean)
print(metrics.r2_score(y_test,pred1))
print(metrics.mean_absolute_error(y_test,pred1))
print(np.sqrt(metrics.mean_squared_error(y_test,pred1)))
print(np.median((np.abs(np.subtract(y_test,pred1)/ y_test))) * 100)
```
